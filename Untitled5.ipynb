```python
data['DATE'] = pd.to_datetime(data['DATE'], format='%d/%m/%Y')
X_train = data[['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET']]
y_train = data['WATER YIELD']
```

```python
# Train the Lasso Regression model
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
```

```python
# Load the dataset from 2024 to 2030 with empty 'WATER YIELD'
data_to_predict = pd.read_csv('/content/finaltes.csv')

# Preprocess the data to be predicted
# Convert the 'DATE' column to datetime format
data_to_predict['DATE'] = pd.to_datetime(data_to_predict['DATE'], format='%d-%m-%Y')

X_to_predict = data_to_predict[['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET']]

# Predict water yield for each month from 2024 to 2030 using the trained Lasso regression model
predicted_water_yield = []

for year in range(2024, 2031):
    for month in range(1, 13):
        # Filter data for the specific year and month
        X_month = X_to_predict[(data_to_predict['DATE'].dt.year == year) & (data_to_predict['DATE'].dt.month == month)]

        # Predict water yield for the month
        if not X_month.empty:
            y_pred_month = lasso.predict(X_month)
            predicted_water_yield.append((year, month, y_pred_month[0]))
        else:
            predicted_water_yield.append((year, month, None))

# Print the predictions
print("Predicted Water Yield for each month from 2024 to 2030:")
for prediction in predicted_water_yield:
    year, month, pred = prediction
    if pred is not None:
        print(f"Year: {year}, Month: {month}, Predicted Water Yield: {pred}")
    else:
        print(f"Year: {year}, Month: {month}, No data available for prediction")
```

```python
 yearly_predictions = {}
 for prediction in predicted_water_yield:
    year, month, pred = prediction
    if pred is not None:
        if year not in yearly_predictions:
            yearly_predictions[year] = []
        yearly_predictions[year].append(pred)
 # Calculate yearly average water yield
 print("\nYearly Predicted Water Yield:")
 for year, predictions in yearly_predictions.items():
    if predictions:
        yearly_avg = sum(predictions)
        print(f"Year: {year},Predicted Water Yield: {yearly_avg}")
    else:
        print(f"Year: {year}, No data available for prediction")
```

```python
 import matplotlib.pyplot as plt
 # Plot yearly predictions from 2024 to 2030
 years = list(yearly_predictions.keys())
 predicted_yield = [sum(yearly_predictions[year]) for year in years]
 plt.figure(figsize=(10, 6))
 plt.plot(years, predicted_yield, marker='o', color='b', linestyle='-')
 plt.title('Yearly Predicted Water Yield (2024-2030)')
 plt.xlabel('Year')
 plt.ylabel('Predicted Water Yield (cubic metres)')
 plt.grid(True)
 plt.show()
```

```python
 import pandas as pd
 import numpy as np
 from sklearn.linear_model import Lasso, Ridge, ElasticNet
 from sklearn.metrics import mean_squared_error, r2_score
 import matplotlib.pyplot as plt
 # Load dataset
 data = pd.read_csv('/content/finaltra.csv')
 # Preprocess the data
 data['DATE'] = pd.to_datetime(data['DATE'], format='%d/%m/%Y')
 # Split into train and test sets
 train_data = data[data['DATE'].dt.year <= 2015]
 test_data = data[data['DATE'].dt.year >= 2016]
 X_train = train_data[['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET']]
 y_train = train_data['WATER YIELD']
 X_test = test_data[['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET']]
 y_test = test_data['WATER YIELD']
 # Train the models
 lasso = Lasso(alpha=0.1)
 lasso.fit(X_train, y_train)
 ridge = Ridge(alpha=0.1)
 ridge.fit(X_train, y_train)
 elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
 elastic_net.fit(X_train, y_train)
 # Calculate metrics for each model
 lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso.predict(X_test)))
 ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge.predict(X_test)))
 elastic_net_rmse = np.sqrt(mean_squared_error(y_test, elastic_net.predict(X_test)))
 lasso_mse = mean_squared_error(y_test, lasso.predict(X_test))
 ridge_mse = mean_squared_error(y_test, ridge.predict(X_test))
 elastic_net_mse = mean_squared_error(y_test, elastic_net.predict(X_test))
 lasso_r_squared = r2_score(y_test, lasso.predict(X_test))
 ridge_r_squared = r2_score(y_test, ridge.predict(X_test))
 elastic_net_r_squared = r2_score(y_test, elastic_net.predict(X_test))
 print("Performance Metrics Comparison:")
 print("RMSE - Lasso:", lasso_rmse)
 print("RMSE - Ridge:", ridge_rmse)
 print("RMSE - ElasticNet:", elastic_net_rmse)
 print("MSE - Lasso:", lasso_mse)
 print("MSE - Ridge:", ridge_mse)
 print("MSE - ElasticNet:", elastic_net_mse)
 # Plotting
 plt.figure(figsize=(20, 10))
 # RMSE Comparison
 plt.subplot(2, 2, 1)
 plt.bar(['Lasso', 'Ridge', 'ElasticNet'], [lasso_rmse, ridge_rmse, elastic_net_rmse])
 plt.title('RMSE Comparison')
 plt.xlabel('Regression Models')
 plt.ylabel('RMSE')
 # MSE Comparison
 plt.subplot(2, 2, 2)
 plt.bar(['Lasso', 'Ridge', 'ElasticNet'], [lasso_mse, ridge_mse, elastic_net_mse])
 plt.title('MSE Comparison')
 plt.xlabel('Regression Models')
 plt.ylabel('MSE')
 plt.tight_layout()
 plt.show()
```

```python
import gradio as gr
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load and preprocess data (assuming 'finaltra.csv' is in the same directory)
try:
    data = pd.read_csv('finaltra.csv')
    data['DATE'] = pd.to_datetime(data['DATE'], format='%d/%m/%Y')

    # Split into train and test sets
    train_data = data[data['DATE'].dt.year <= 2015]
    test_data = data[data['DATE'].dt.year >= 2016]

    X_train = train_data[['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET']]
    y_train = train_data['WATER YIELD']
    X_test = test_data[['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET']]
    y_test = test_data['WATER YIELD']

    # Train the models
    lasso = Lasso(alpha=0.1)
    lasso.fit(X_train, y_train)

    ridge = Ridge(alpha=0.1)
    ridge.fit(X_train, y_train)

    elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
    elastic_net.fit(X_train, y_train)

except FileNotFoundError:
    print("Ensure 'finaltra.csv' is in the correct directory.")
    # Create dummy models if the file is not found, to allow the Gradio interface to launch
    class DummyModel:
        def predict(self, X):
            return np.zeros(X.shape[0])
    lasso = DummyModel()
    ridge = DummyModel()
    elastic_net = DummyModel()
    X_test = pd.DataFrame()
    y_test = pd.Series()


# Prediction function
def predict_water_yield(model_name, prec, surq, latq, gwq, percolate, sw, et, pet):
    """
    Predicts water yield based on the selected model and input features.
    """
    input_data = pd.DataFrame([[prec, surq, latq, gwq, percolate, sw, et, pet]],
                              columns=['PREC', 'SURQ', 'LATQ', 'GWQ', 'PERCOLATE',  'SW', 'ET', 'PET'])

    if model_name == "Lasso":
        prediction = lasso.predict(input_data)[0]
    elif model_name == "Ridge":
        prediction = ridge.predict(input_data)[0]
    else: # ElasticNet
        prediction = elastic_net.predict(input_data)[0]

    return round(prediction, 4)

# Function to generate and display comparison plots
def get_comparison_plots():
    """
    Generates and returns comparison plots for the models.
    """
    if not X_test.empty and not y_test.empty:
        # Calculate metrics
        lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso.predict(X_test)))
        ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge.predict(X_test)))
        elastic_net_rmse = np.sqrt(mean_squared_error(y_test, elastic_net.predict(X_test)))

        lasso_mse = mean_squared_error(y_test, lasso.predict(X_test))
        ridge_mse = mean_squared_error(y_test, ridge.predict(X_test))
        elastic_net_mse = mean_squared_error(y_test, elastic_net.predict(X_test))

        # Create plots
        fig, axs = plt.subplots(2, 2, figsize=(15, 10))

        # RMSE Comparison
        axs[0, 0].bar(['Lasso', 'Ridge', 'ElasticNet'], [lasso_rmse, ridge_rmse, elastic_net_rmse])
        axs[0, 0].set_title('RMSE Comparison')
        axs[0, 0].set_xlabel('Regression Models')
        axs[0, 0].set_ylabel('RMSE')

        # MSE Comparison
        axs[0, 1].bar(['Lasso', 'Ridge', 'ElasticNet'], [lasso_mse, ridge_mse, elastic_net_mse])
        axs[0, 1].set_title('MSE Comparison')
        axs[0, 1].set_xlabel('Regression Models')
        axs[0, 1].set_ylabel('MSE')

        # Residual Curves
        axs[1, 0].plot(y_test - lasso.predict(X_test), label='Lasso Residuals')
        axs[1, 0].plot(y_test - ridge.predict(X_test), label='Ridge Residuals')
        axs[1, 0].plot(y_test - elastic_net.predict(X_test), label='ElasticNet Residuals')
        axs[1, 0].set_xlabel('Water yield (cubic metres)')
        axs[1, 0].set_ylabel('Residuals')
        axs[1, 0].set_title('Residual Curves')
        axs[1, 0].legend()

        # Hide the fourth subplot
        axs[1, 1].set_visible(False)


        plt.tight_layout()
        return fig
    else:
        return None


# Create the Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Water Yield Prediction")
    gr.Markdown("Enter the values for the features to predict the water yield.")

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Input Features")
            model_name = gr.Dropdown(["Lasso", "Ridge", "ElasticNet"], label="Select Model", value="Lasso")
            prec = gr.Number(label="Precipitation (PREC)")
            surq = gr.Number(label="Surface Runoff (SURQ)")
            latq = gr.Number(label="Lateral Flow (LATQ)")
            gwq = gr.Number(label="Groundwater Flow (GWQ)")
            percolate = gr.Number(label="Percolation (PERCOLATE)")
            sw = gr.Number(label="Soil Water (SW)")
            et = gr.Number(label="Evapotranspiration (ET)")
            pet = gr.Number(label="Potential Evapotranspiration (PET)")
            predict_btn = gr.Button("Predict")

        with gr.Column():
            gr.Markdown("### Prediction")
            predicted_yield = gr.Number(label="Predicted Water Yield")

    predict_btn.click(predict_water_yield,
                      inputs=[model_name, prec, surq, latq, gwq, percolate, sw, et, pet],
                      outputs=predicted_yield)

    gr.Markdown("---")
    gr.Markdown("## Model Performance Comparison")

    with gr.Row():
        plot_output = gr.Plot()

    demo.load(get_comparison_plots, inputs=None, outputs=plot_output)


if __name__ == "__main__":
  demo.launch()
```
